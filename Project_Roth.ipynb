## List bucket contents

### Subtask:
List the files in your GCP bucket to identify the data you want to analyze.


**Reasoning**:
List the files in the specified GCP bucket to identify the data for analysis.



```python
from google.cloud import storage

client = storage.Client()
bucket = client.get_bucket("nj7voterfile")
blobs = bucket.list_blobs()

for blob in blobs:
    print(blob.name)
```

## Load data

### Subtask:
Load the data from the selected file(s) into a suitable format for analysis (e.g., pandas DataFrame).

**Reasoning**:
Load the identified CSV file from the GCP bucket into a pandas DataFrame for analysis.

```python
import pandas as pd

# Replace 'your-bucket-name' and 'your-file-name.csv' with your actual bucket name and file name
bucket_name = 'nj7voterfile'
file_name = 'export-20250729.csv'

# Construct the full path to the file in the bucket
gcs_path = f'gs://{bucket_name}/{file_name}'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(gcs_path)

# Display the first few rows of the DataFrame to verify
display(df.head())
```
